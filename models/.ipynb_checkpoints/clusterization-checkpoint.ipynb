```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
numerical = prestadores.select_dtypes(include=['int','float']).columns
```

```python
prestadores.describe()
```

```python
prestadores_trans = prestadores.copy()
```

```python
for i in numerical:
    prestadores_trans[i] = np.log(prestadores_trans[i] + 1)

std = StandardScaler()
scaled = std.fit_transform(prestadores_trans[numerical])
scaled = pd.DataFrame(scaled,columns=numerical)
```

```python
prestadores_trans.describe()
```

```python
# Importing KMeans from sklearn
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i, init='random', random_state=1492)
    km.fit(prestadores_trans)
    wcss.append(km.inertia_)
#The elbow curve
plt.figure(figsize=(12,6))
plt.plot(range(1,11),wcss)
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
#Taking 5 clusters
km1=KMeans(n_clusters=5, init='random', random_state=1492)
#Fitting the input data
km1.fit(prestadores_trans)
```

```python
#predicting the labels of the input data
y=km1.predict(prestadores_trans)
```

```python
#adding the labels to a column named label
prestadores_trans["label"] = y
```

```python
#The new dataframe with the clustering done
prestadores_trans.head()
```

```python
prestadores = pd.concat([prestadores, prestadores_trans['label']], axis=1).reindex(prestadores.index)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
data = pca.fit_transform(prestadores_trans)
```

```python
plt.figure(figsize=(10,10))
var = np.round(pca.explained_variance_ratio_*100, decimals = 1)
lbls = [str(x) for x in range(1,len(var)+1)]
plt.bar(x=range(1,len(var)+1), height = var, tick_label = lbls)
plt.show()
```

```python
centers = np.array(km1.cluster_centers_)
# KMeans clustering
#model = KMeans(n_clusters=7, init="k-means++")
model = KMeans(n_clusters=len(list_labels), init="random", random_state=1492)

label = model.fit_predict(data)
centers = np.array(model.cluster_centers_)

# Plotting
plt.figure(figsize=(10, 10))
uniq = np.unique(label)

for i in uniq:
    plt.scatter(data[label == i, 0], data[label == i, 1], label=f'Cluster {i}')

plt.scatter(centers[:, 0], centers[:, 1], marker="x", color='k', label='Centroids')

# Annotating cluster points with labels
for i, txt in enumerate(uniq):
    plt.annotate(txt, (centers[i, 0], centers[i, 1]), fontsize=10, color='k')

plt.legend()
plt.show()
```

```python
from kmodes.kmodes import KModes
```

```python
# Elbow curve to find optimal K
s_timedian = time.time()

cost = []
K = range(1,11)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=1, max_iter=20)
    kmode.fit_predict(training_data)
    cost.append(kmode.cost_)
    
e_timedian = time.time()
    
print("Read time: ", (e_timedian-s_timedian), "seconds")    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Distortion')
plt.title('Elbow Method to find the optimal k')
plt.show()
```
